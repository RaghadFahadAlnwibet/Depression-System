# Depression Detection using RNN.

```python
# Base Model (Model 1)

from tensorflow.keras.layers import SimpleRNN
def depression_model():
    sequence_input = Input(shape=(maxlength,), dtype='int32')# input layer
    embedding_layer = Embedding(len(word_index) + 1, # embedding layer
                               EMBEDDING_DIM,
                               weights = [embedding_matrix],
                               input_length = maxlength,
                               trainable=False,
                               name = 'embeddings')
    embedded_sequences = embedding_layer(sequence_input)# passes the input sequences through the embedding layer, convert the 2D input tensor into a 3D tensor
    x = SimpleRNN(60, return_sequences=True,name='SimpleRNN_layer')(embedded_sequences)
    x = GlobalMaxPool1D()(x)
    x = Dropout(0.3)(x)
    x = Dense(50, activation="relu")(x)
    x = Dropout(0.3)(x)
    preds = Dense(1, activation="sigmoid")(x)

    model = Model(sequence_input, preds)
    model.compile(loss = 'binary_crossentropy',
             optimizer='adam', # optimization algorithm to update network weights iteratively based on training data
             metrics = ['accuracy'])

    return model
model = depression_model()
model.summary()
```

# Hyperparameter Tuning

```python
import tensorflow as tf
!pip install -q -U keras-tuner
```

```python
from tensorflow.keras.layers import SimpleRNN
def model_builder(hp):
    sequence_input = Input(shape=(maxlength,), dtype='int32')
    embedding_layer = Embedding(len(word_index) + 1, EMBEDDING_DIM, weights=[embedding_matrix],
                                input_length=maxlength, trainable=False, name='embeddings')
    embedded_sequences = embedding_layer(sequence_input)

    # Hyperparameters for the RNN layer
    hp_units = hp.Int('SimpleRNN_units', min_value=32, max_value=256, step=32)
    hp_dropout = hp.Float('SimpleRNN_dropout', min_value=0.0, max_value=0.5, step=0.1)
    # start
    # RNN layer with hyperparameters
    x = SimpleRNN(hp_units, return_sequences=True, dropout=hp_dropout, name='RNN_layer')(embedded_sequences)
    x = GlobalMaxPool1D()(x)
    # Hyperparameters for the first dense layer
    hp_dropout_dense = hp.Float('dense_dropout', min_value=0.0, max_value=0.5, step=0.1)
    hp_units_dense1 = hp.Int('dense_1_units', min_value=32, max_value=512, step=32)

    # Dense layer with hyperparameters
    x = Dense(hp_units_dense1, activation='relu')(x)
    x = Dropout(hp_dropout_dense)(x)

    # Output layer
    hp_dropout_dense = hp.Float('dense_dropout', min_value=0.0, max_value=0.5, step=0.1)
    preds = Dense(1, activation='sigmoid')(x)

    # Create the model
    model = Model(sequence_input, preds)
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(optimizer=Adam(learning_rate=hp_learning_rate),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model
```

```python
import keras_tuner as kt

tuner = kt.Hyperband(model_builder,
                     objective='val_accuracy',
                     max_epochs=10,
                     factor=3,
                     directory='my_dir',
                     project_name='intro_to_kt')
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
```

```python
tuner.search(X_train, y_train, epochs=10, validation_split=0.2, callbacks=[stop_early])
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
model = tuner.hypermodel.build(best_hps)
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2,
                    callbacks=[stop_early])
```

#Hyperparameter tunning

```python
from tensorflow.keras.regularizers import l2

def model_builder(hp):
    sequence_input = Input(shape=(maxlength,), dtype='int32')
    embedding_layer = Embedding(len(word_index) + 1, EMBEDDING_DIM, weights=[embedding_matrix], input_length=maxlength, trainable=False, name='embeddings')
    embedded_sequences = embedding_layer(sequence_input)

    # Hyperparameters for the LSTM layer
    hp_units = hp.Int('lstm_units', min_value=32, max_value=256, step=32)
    hp_dropout = hp.Float('lstm_dropout', min_value=0.0, max_value=0.5, step=0.1)

    # LSTM layer with hyperparameters
    x = LSTM(hp_units, return_sequences=True, dropout=hp_dropout, name='lstm_layer')(embedded_sequences)
    x = GlobalMaxPool1D()(x)

    # Hyperparameters for the first dense layer
    hp_dropout_dense = hp.Float('dense_dropout', min_value=0.0, max_value=0.5, step=0.1)
    hp_units_dense1 = hp.Int('dense_1_units', min_value=32, max_value=512, step=32)

    # Hyperparameter for L2 regularization
    hp_lambda = hp.Choice('lambda', values=[1e-4, 1e-3, 1e-2])

    # Dense layer with hyperparameters
    x = Dense(hp_units_dense1, activation='relu', kernel_regularizer=l2(hp_lambda))(x)
    x = Dropout(hp_dropout_dense)(x)

    # Output layer
    hp_dropout_dense = hp.Float('dense_dropout', min_value=0.0, max_value=0.5, step=0.1)
    preds = Dense(1, activation='sigmoid')(x)

    # Create the model
    model = Model(sequence_input, preds)
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(optimizer=Adam(learning_rate=hp_learning_rate), loss='binary_crossentropy', metrics=['accuracy'])

    return model
```
